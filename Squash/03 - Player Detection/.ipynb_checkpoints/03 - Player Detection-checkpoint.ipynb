```python
import sys
from keras.models import load_model
import cv2 as cv2
from datetime import datetime
import numpy as np
from imutils.video import VideoStream, FPS
import imutils
import time

path = '/media/sf_Google_Drive/Data Science/Deep Learning/Jason Brownlee - Deep Learning for Computer Vision/code/chapter_24 - YOLO/'
sys.path.insert(0, path + 'keras-yolo3-master/')
import yolo3_one_file_to_detect_them_all as yolo3
```

```python
path_dataset = '/media/sf_Google_Drive/LudisAI/00 - Datasets/'
```

### Instanciar el modelo

```python
yolo_model = load_model(path + 'yolo_model.h5')
```

```python
net_h, net_w = 416, 416
obj_thresh, nms_thresh = 0.5, 0.45
anchors = [[116,90,  156,198,  373,326],  [30,61, 62,45,  59,119], [10,13,  16,30,  33,23]]
labels = ["person"]
```

### Image Player Detection

```python
# preprocess the image
image_path = path_dataset + 'squash/_MG_1970.JPG'
image = cv2.imread(image_path)
image_h, image_w, _ = image.shape
new_image = yolo3.preprocess_input(image, net_h, net_w)

# run the prediction
yolos = yolo_model.predict(new_image)
boxes = []

for i in range(len(yolos)):
    # decode the output of the network
    boxes += yolo3.decode_netout(yolos[i][0], anchors[i], obj_thresh, nms_thresh, net_h, net_w)
    
# correct the sizes of the bounding boxes
yolo3.correct_yolo_boxes(boxes, image_h, image_w, net_h, net_w)

# suppress non-maximal boxes
yolo3.do_nms(boxes, nms_thresh)  

# draw bounding boxes on the image using labels
yolo3.draw_boxes(image, boxes, labels, obj_thresh) 

# write the image with bounding boxes to file
cv2.imwrite(image_path[:-4] + '_detected' + image_path[-4:], (image).astype('uint8')) 
```

### Video Player Detection

```python
start_time = datetime.now()

# Using OpenCV to initialize the webcam
video_path = path_dataset + 'video_test/squash_beginner.mp4'
output_path = path_dataset + 'video_test/squash_beginner_yolo.mp4'

cap = cv2.VideoCapture(video_path)
video_FourCC = int(cap.get(cv2.CAP_PROP_FOURCC))
video_fps    = cap.get(cv2.CAP_PROP_FPS)
video_size   = (int(cap.get(cv2.CAP_PROP_FRAME_WIDTH)), int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT)))

writer = cv2.VideoWriter(output_path, video_FourCC, video_fps, video_size)

q_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
deciles = [int(x) for x in np.linspace(0,1,11)*q_frames]
frame_number = 0

while True:
    ret, frame = cap.read()
    frame_number += 1
    
    if ret == True:
        # preprocess the image
        image_h, image_w, _ = frame.shape
        new_image = yolo3.preprocess_input(frame, net_h, net_w)

        # run the prediction
        yolos = yolo_model.predict(new_image)
        boxes = []

        for i in range(len(yolos)):
            # decode the output of the network
            boxes += yolo3.decode_netout(yolos[i][0], anchors[i], obj_thresh, nms_thresh, net_h, net_w)

        # correct the sizes of the bounding boxes
        yolo3.correct_yolo_boxes(boxes, image_h, image_w, net_h, net_w)

        # suppress non-maximal boxes
        yolo3.do_nms(boxes, nms_thresh) 

        # draw bounding boxes on the image using labels
        yolo3.draw_boxes(frame, boxes, labels, obj_thresh) 

        if frame_number in deciles:
            print('Processed frames: {} de {} ({}%)'.format(frame_number, q_frames, np.round(frame_number*100/q_frames, 0)))


        # write the output frame to disk
        writer.write(frame)

        #cv2.imshow("Objet Detection YOLO", frame)

        if frame_number == q_frames:
            break
        if cv2.waitKey(1) == 13: #13 is the Enter Key
            break

# release the file pointers
print("[INFO] cleaning up...")
writer.release()
cap.release()
cv2.destroyAllWindows()

duration = datetime.now()-start_time
print('Duracion: ', duration)
```
