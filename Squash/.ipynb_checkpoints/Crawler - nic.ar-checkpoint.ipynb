# Crawler Final

```python
browser = webdriver.Chrome('E:/Crawler NIC/chromedriver.exe') 
```

```python
clientes_web = pd.DataFrame(columns=('Dominio','Extension','CUIT','fc_alta','fc_vencimiento'))

#clientes_particion = clientes_crawler[clientes_crawler.particion != 100].reset_index().drop(columns='index').copy()

for i in range(len(clientes)):
    #time.sleep(1)
    browser.get('https://GOOGLE.COM.AR/')
    browser.find_element_by_xpath('//*[@id="tsf"]/div[2]/div/div[1]/div/div[1]/input').clear()
    browser.find_element_by_xpath('//*[@id="tsf"]/div[2]/div/div[1]/div/div[1]/input').send_keys(clientes.Razon_Social[i])
    time.sleep(1)
    try:
        browser.find_element_by_xpath('//*[@id="tsf"]/div[2]/div/div[3]/center/input[1]').click()
    except:
        try:
            browser.find_element_by_xpath('//*[@id="tsf"]/div[2]/div/div[2]/div[2]/div/center/input[1]').click()
        except:
            try:
                browser.find_element_by_xpath('//*[@id="tsf"]/div[2]/div/div[2]/div[2]/ul/li/div[1]/div').click()
                browser.find_element_by_xpath('//*[@id="tsf"]/div[2]/div/div[3]/center/input[1]').click()
            except:
                continue
            
    webs = []

    try:
        webs.append(browser.find_element_by_xpath('//*[@id="rso"]/div[1]/div/div[1]/div/div/div[1]/a/div/cite').text)
    except:
        pass
    try:
        webs.append(browser.find_element_by_xpath('//*[@id="rso"]/div[1]/div/div[2]/div/div/div[1]/a/div/cite').text)
    except:
        pass
    try:
        webs.append(browser.find_element_by_xpath('//*[@id="rso"]/div[1]/div/div[3]/div/div/div[1]/a/div/cite').text)
    except:
        pass
    try:
        webs.append(browser.find_element_by_xpath('//*[@id="rso"]/div[1]/div/div[4]/div/div/div[1]/a/div/cite').text)
    except:
        pass
    try:
        webs.append(browser.find_element_by_xpath('//*[@id="rso"]/div[1]/div/div[5]/div/div/div[1]/a/div/cite').text)
    except:
        pass
    try:
        webs.append(browser.find_element_by_xpath('//*[@id="rso"]/div[1]/div/div[6]/div/div/div[1]/a/div/cite').text)
    except:
        pass
    try:
        webs.append(browser.find_element_by_xpath('//*[@id="rso"]/div[1]/div/div[7]/div/div/div[1]/a/div/cite').text)
    except:
        pass


    df_web = pd.DataFrame(columns=('Dominio','Extension','CUIT','fc_alta','fc_vencimiento'))

    for i in range(len(webs)):
        if "www" in webs[i].split('.')[0]: 
            dom = webs[i].split('.')[1]
            try:
                ext = (webs[i].split('.')[2]+'.'+webs[i].split('.')[3]).split('/')[0]
            except IndexError:
                try:
                    ext = (webs[i].split('.')[2]).split('/')[0]   
                except IndexError:
                    ext = (webs[i].split('.')[1]).split('/')[0]   
        elif "http" in webs[i]:
            dom = webs[i].split('/')[2].split('.')[0]
        else:
            dom = webs[i].split('.')[0]
            try:
                ext = (webs[i].split('.')[1]+'.'+webs[i].split('.')[2]).split('/')[0]
            except IndexError:
                try:
                    ext = (webs[i].split('.')[1]).split('/')[0]   
                except IndexError:
                    ext = (webs[i].split('.')[0]).split('/')[0]  

        df_web = df_web.set_value(len(df_web),['Dominio','Extension'],[dom,ext])
        df_web = df_web[df_web.Dominio != 'dateas'].copy()
        df_web = df_web[df_web.Dominio != 'cuitonline'].copy()
        df_web = df_web[df_web.Dominio != 'buscardatos'].copy()
        df_web = df_web[df_web.Dominio != 'twitter'].copy()
        df_web = df_web[df_web.Dominio != 'facebook'].copy()
        df_web = df_web[df_web.Dominio != 'trade'].copy()
        df_web = df_web[df_web.Dominio != 'infobel'].copy()
        df_web = df_web[df_web.Dominio != 'abctelefonos'].copy()
        df_web = df_web[df_web.Dominio != 'paginasamarillas'].copy()
        df_web = df_web[df_web.Dominio != 'es-la'].copy()
        df_web = df_web[df_web.Dominio != 'paginasblancas'].copy()
        df_web = df_web[df_web.Dominio != 'yelp'].copy()
        df_web = df_web[df_web.Dominio != 'es'].copy()
        df_web = df_web[df_web.Dominio != 'ar'].copy()
        df_web = df_web[df_web.Dominio != 'es-la'].copy()
        df_web = df_web[df_web.Dominio != 'linkedin'].copy()
        df_web = df_web[df_web.Dominio != 'buenos-aires'].copy()
        df_web = df_web[df_web.Dominio != 'webdedatos'].copy()
        df_web = df_web[df_web.Dominio != 'elnuevoherald'].copy()
        df_web = df_web[df_web.Dominio != 'argentino'].copy()
        df_web = df_web[df_web.Dominio != 'telefonos'].copy()
    df_web = df_web.drop_duplicates().reset_index().drop(columns='index').copy()

    try:
        for i in range(len(df_web)):
            browser.get('https://nic.ar/')
            time.sleep(2)
            #browser.find_element_by_xpath('//*[@id="edit-txtbuscar"]').clear()
            browser.find_element_by_xpath('//*[@id="edit-txtbuscar"]').send_keys(df_web.Dominio[i])
            if "com" in df_web.Extension[i]:
                browser.find_element_by_xpath('//*[@id="cmbZonas"]').send_keys('.com.ar')
            elif "org" in df_web.Extension[i]:
                browser.find_element_by_xpath('//*[@id="cmbZonas"]').send_keys('.org.ar')
            elif "net" in df_web.Extension[i]:
                browser.find_element_by_xpath('//*[@id="cmbZonas"]').send_keys('.net.ar')
            elif "gob" in df_web.Extension[i]:
                browser.find_element_by_xpath('//*[@id="cmbZonas"]').send_keys('.gob.ar')
            elif "tur" in df_web.Extension[i]:
                browser.find_element_by_xpath('//*[@id="cmbZonas"]').send_keys('.tur.ar')

            browser.find_element_by_xpath('//*[@id="btn-consultar-block-submit"]').click()
            try:
                if browser.find_element_by_xpath('/html/body/div/div[3]/div/section/div/div/p').text == '¡Sí! El dominio está disponible':
                    cuit = None
                    fc_alta = None
                    fc_vencimiento = None
                else:
                    cuit = browser.find_element_by_xpath('/html/body/div[1]/div[3]/div/section/div/div/div[4]/table[1]/tbody/tr[2]/td').text.split(':')[1]
                    fc_alta = browser.find_element_by_xpath('/html/body/div[1]/div[3]/div/section/div/div/div[4]/table[1]/tbody/tr[3]/td').text.split(':')[1]
                    fc_vencimiento = browser.find_element_by_xpath('/html/body/div[1]/div[3]/div/section/div/div/div[4]/table[1]/tbody/tr[4]/td').text.split(':')[1]
            except:
                try:
                    cuit = browser.find_element_by_xpath('/html/body/div[1]/div[3]/div/section/div/div[2]/table/tbody/tr[2]/td').text.split(':')[1]
                    fc_alta = browser.find_element_by_xpath('/html/body/div[1]/div[3]/div/section/div/div[2]/table/tbody/tr[3]/td').text.split(':')[1]
                    fc_vencimiento = browser.find_element_by_xpath('/html/body/div[1]/div[3]/div/section/div/div[2]/table/tbody/tr[4]/td').text.split(':')[1]
                except:
                    pass

            df_web['CUIT'][i] = cuit
            df_web['fc_alta'][i] = fc_alta
            df_web['fc_vencimiento'][i] = fc_vencimiento
            #browser.find_element_by_xpath('//*[@id="navbar"]/div[1]/div[1]/div/a/img').click()

        clientes_web = clientes_web.append(df_web)
        clientes_web = clientes_web.drop_duplicates().reset_index().drop(columns='index').copy()
    except:
        continue
```
