### Object Detection and Object Tracking

**HSV (Hue, Saturation and Value)**

**Hue (Matriz):** color components (0-360)   
**Saturation:** amount of color (depth of the pigment)(0-100%)   
**Value:** brightness of the color (0-100%) 



```python
def nothing(x):
    pass

cap = cv2.VideoCapture(0)

cv2.namedWindow('Tracking')

# Creo los trackbars para HSV lower y upper
cv2.createTrackbar('LH', 'Tracking', 0, 255, nothing)
cv2.createTrackbar('LS', 'Tracking', 0, 255, nothing)
cv2.createTrackbar('LV', 'Tracking', 0, 255, nothing)
cv2.createTrackbar('UH', 'Tracking', 255, 255, nothing)
cv2.createTrackbar('US', 'Tracking', 255, 255, nothing)
cv2.createTrackbar('UV', 'Tracking', 255, 255, nothing)


while True:
    #frame = cv2.imread(path+'smarties.png')
    
    _, frame = cap.read()
    
    hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)
    
    l_h = cv2.getTrackbarPos('LH', 'Tracking')
    l_s = cv2.getTrackbarPos('LS', 'Tracking')
    l_v= cv2.getTrackbarPos('LV', 'Tracking')
    
    u_h = cv2.getTrackbarPos('UH', 'Tracking')
    u_s = cv2.getTrackbarPos('US', 'Tracking')
    u_v= cv2.getTrackbarPos('UV', 'Tracking')
    
    l_b = np.array([l_h, l_s, l_v])
    u_b = np.array([u_h, u_s, u_v])
    
    # Creo una mascara con la que se define el color a detectar
    mask = cv2.inRange(hsv, l_b, u_b)
    
    # Se une 2 veces la misma imagen aplicando la mascara
    res = cv2.bitwise_and(frame, frame, mask=mask)
    
    cv2.imshow('frame', res)
    
    k = cv2.waitKey(1) & 0xFF
    if k == 27:
        break

cap.release()
cv2.destroyAllWindows()
```

### Morphological Transformations

**Kernel:** indica como transformar cualquier pixel dado, combinandolo con diferentes pixeles vecinos

```python
img = cv2.imread(path+'smarties.png', 0)

# Las transformaciones se realizan sobre imagenes binarias, por lo que se aplica una mascara binaria a la imagen
_, mask = cv2.threshold(img, thresh=220, maxval=255, type=cv2.THRESH_BINARY_INV)

# Para sacar los puntitos negros que quedan en la imagen binaria, se usa dilation
kernal = np.ones((2,2), np.uint8)
dilation = cv2.dilate(mask, kernel=kernal, iterations=2)

# Erosion (erosionan los contornos, asignandoles 0 (negro))
erosion = cv2.erode(mask, kernel=kernal, iterations=1)

# Opening (Erosion + Dilation)
opening = cv2.morphologyEx(mask, op=cv2.MORPH_OPEN, kernel=kernal)

# Closing (Dilation + Erosion)
closing = cv2.morphologyEx(mask, op=cv2.MORPH_CLOSE, kernel=kernal)

# Gradient (Da la diferencia entre Dilation y Erosion)
gradient = cv2.morphologyEx(mask, op=cv2.MORPH_GRADIENT, kernel=kernal)

# TopHat (Da la diferencia entre Input Image y su Opening)
tophat = cv2.morphologyEx(mask, op=cv2.MORPH_TOPHAT, kernel=kernal)

titles = ['image', 'mask', 'dilation', 'erosion', 'opening', 'closing', 'gradient', 'tophat']
images = [img, mask, dilation, erosion, opening, closing, gradient, tophat]

for i in range(len(images)):
    #plt.figure(figsize=(10,10))
    plt.subplot(2,4,i+1), plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.xticks([]), plt.yticks([])

plt.show()
```

### Smoothing Images

**FILTROS**

**Homogeneous:** cada pixel output es la media de los pixeles vecinos pertenecientes al kernel

K = $\frac{1}{Kwidth*Kheight}$ * OneMatrix

Ejemplo: filtro de 5x5 quedaria $\frac{1}{25}$ * OneMatrix5x5

**Low-Pass (LPF):** ayudan a remover ruido, borroneando la imagen

- **Gaussian:** se usa un kernel ponderado (da mas peso a los valores internos del kernel - desde el medio hacia afuera)

- **Median:** cada pixel output es la mediana de los pixeles vecinos pertenecientes al kernel (sirve para eliminar el efecto "salt and pepper"

**High-Pass (HPF):** ayudan a detectar bordes en la imagen

- **Bilateral:** sirve para preservar los bordes

```python
img = cv2.imread(path+'lena.jpg', 1)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Homogeneous filter
kernel = np.ones((5,5), np.float32) / 25
dst = cv2.filter2D(img, ddepth=-1, kernel=kernel)

# Low-Pass filter
blur = cv2.blur(img, ksize=(5,5))

# Gaussian filter
gblur = cv2.GaussianBlur(img, ksize=(5,5), sigmaX=0)

# Median filter
median = cv2.medianBlur(img, ksize=5)

# Bilateral filter
bilateral = cv2.bilateralFilter(img, d=9, sigmaColor=75, sigmaSpace=75)

titles = ['image', '2D Convolution', 'Blur', 'Gaussian Blur', 'Median', "Bilateral"]
images = [img, dst, blur, gblur, median, bilateral]

for i in range(len(images)): 
    plt.figure(figsize=(20,10))
    plt.subplot(3,2,i+1), plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.xticks([]), plt.yticks([])

plt.show()
```

### Image Gradients and Edge Detection

El gradiente es la direccion en la cual se intensifica el color

```python
img = cv2.imread(path+'messi5.jpg', 0)

# Laplacian Gradient
lap = cv2.Laplacian(img, cv2.CV_64F, ksize=3)
lap = np.uint8(np.absolute(lap))

# Sobel Gradient
sobelX = cv2.Sobel(img, cv2.CV_64F, dx=1, dy=0) # Deriva sobre X
sobelY = cv2.Sobel(img, cv2.CV_64F, dx=0, dy=1) # Deriva sobre Y

sobelX = np.uint8(np.absolute(sobelX))
sobelY = np.uint8(np.absolute(sobelY))

sobelCombined = cv2.bitwise_or(sobelX, sobelY)


canny = cv2.Canny(img, 100,200)
titles = ['image', 'Laplacian', 'SobelX', 'SobelY', 'sobelCombined', 'Canny']
images = [img, lap, sobelX, sobelY, sobelCombined, canny]

for i in range(len(images)): 
    plt.figure(figsize=(20,20))
    plt.subplot(3,2,i+1), plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.xticks([]), plt.yticks([])

plt.show()
```

### Canny Edge Detection

Es un algoritmo para detectar bordes en imagenes (John F. Canny). Esta compuesto por 5 pasos:

**1-** Noise Reduction   
**2-** Gradient Calculation   
**3-** Non-Max Suppression   
**4-** Double Threshold   
**5-** Edge tracking by Hysteresis

```python
img = cv2.imread(path+'lena.jpg', 0)

# Canny
# Thresholds son para el 5to paso
canny = cv2.Canny(img, threshold1=100, threshold2=200)

titles = ['image', 'Canny']
images = [img, canny]

for i in range(len(images)): 
    plt.figure(figsize=(20,20))
    plt.subplot(1,2,i+1), plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.xticks([]), plt.yticks([])

plt.show()
```

### Image Pyramids

Las piramides sirven para mezclar o reconstruir imagenes



```python
img = cv2.imread(path+'lena.jpg', 1)

# Lower Resolution
lr1 = cv2.pyrDown(img)
lr2 = cv2.pyrDown(lr1)

# Increase Resolution (Se pierde calidad por haber reducido la resolucion (se perdio informacion en la convolucion))
hr1 = cv2.pyrUp(lr2)
hr2 = cv2.pyrUp(hr1)

cv2.imshow('Original', img)
cv2.imshow('pyrDown 1', lr1)
cv2.imshow('pyrDown 2', lr2)
cv2.imshow('pyrUp 1', hr1)
cv2.imshow('pyrUp 2', hr2)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img = cv2.imread(path+'lena.jpg', 1)

layer = img.copy()
gp = [layer]

# Gaussian Pyramid
for i in range(6):
    layer = cv2.pyrDown(layer)
    gp.append(layer)
    cv2.imshow(str(i), layer)

# Laplaciand Pyramid (diferencia entre el nivel actual de Gaussian Pyramid con el nivel inmediato superior)
layer = gp[5]
cv2.imshow('Upper level Gaussian Pyramid', layer)
lp = [layer]

for i in range(5,0,-1): # Conteo hacia atras
    gaussian_extended = cv2.pyrUp(gp[i])
    laplacian = cv2.subtract(gp[i-1], gaussian_extended)
    cv2.imshow(str(i), laplacian)    
    
cv2.imshow('Original', img)


cv2.waitKey(0)
cv2.destroyAllWindows()
```

### Blending Images using Pyramids

```python
apple = cv2.imread(path+'apple.jpg', 1)
orange = cv2.imread(path+'orange.jpg', 1)

apple_orange = np.hstack((apple[:,:256], orange[:,256:]))


### Blending
# Gaussian Pyramid Apple
apple_copy = apple.copy()
gp_apple = [apple_copy]

for i in range(6):
    apple_copy = cv2.pyrDown(apple_copy)
    gp_apple.append(apple_copy)
    
# Laplaciand Pyramid Apple
apple_copy = gp_apple[5]
lp_apple = [apple_copy]

for i in range(5,0,-1): # Conteo hacia atras
    gaussian_extended = cv2.pyrUp(gp_apple[i])
    laplacian = cv2.subtract(gp_apple[i-1], gaussian_extended)
    lp_apple.append(laplacian)    

# Gaussian Pyramid Orange
orange_copy = orange.copy()
gp_orange = [orange_copy]

for i in range(6):
    orange_copy = cv2.pyrDown(orange_copy)
    gp_orange.append(orange_copy)
    
# Laplaciand Pyramid Orange
orange_copy = gp_orange[5]
lp_orange = [orange_copy]

for i in range(5,0,-1): # Conteo hacia atras
    gaussian_extended = cv2.pyrUp(gp_orange[i])
    laplacian = cv2.subtract(gp_orange[i-1], gaussian_extended)
    lp_orange.append(laplacian)    

apple_orange_pyramid = []
n = 0
for apple_lap, orange_lap in zip(lp_apple, lp_orange):
    n+=1
    cols, rows, ch = apple_lap.shape
    laplacian = np.hstack((apple_lap[:,0:int(cols/2)], orange_lap[:,int(cols/2):]))
    apple_orange_pyramid.append(laplacian)

apple_orange_reconstruct = apple_orange_pyramid[0]

for i in range(1,6):
    apple_orange_reconstruct = cv2.pyrUp(apple_orange_reconstruct)
    apple_orange_reconstruct = cv2.add(apple_orange_pyramid[i], apple_orange_reconstruct)
    

cv2.imshow('apple', apple)
cv2.imshow('orange', orange)
cv2.imshow('apple_orange', apple_orange)
cv2.imshow('apple_orange_reconstruct ', apple_orange_reconstruct)


cv2.waitKey(0)
cv2.destroyAllWindows()
```

### Find and Draw Contours

Cada contorno es un array de NumPy de coordenadas (x,y) que hacen el contorno del objeto

```python
img = cv2.imread(path+'opencv-logo.png', 1)
imgray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

ret, thresh = cv2.threshold(imgray, thresh = 127, maxval = 255, type = 0)
contours, hierarchy = cv2.findContours(image=thresh, mode=cv2.RETR_TREE, method=cv2.CHAIN_APPROX_NONE)
# contours es una lista de contornos

print('Number of contours: ', str(len(contours)))
print(contours[0])

cv2.drawContours(img, contours=contours, contourIdx=-1, color=(0,255,0), thickness=3)

cv2.imshow('image', img)
cv2.imshow('imgae GRAY', imgray)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

### Motion Detection and Tracking

```python
cap = cv2.VideoCapture(path+'vtest.avi') 

ret, frame1 = cap.read()
ret, frame2 = cap.read()


while (cap.isOpened()):
    diff = cv2.absdiff(frame1, frame2)
    # Lo paso a escala de grises para poder detectar contornos
    gray = cv2.cvtColor(diff, cv2.COLOR_BGR2GRAY)
    blur = cv2.GaussianBlur(gray, ksize=(5,5), sigmaX=0)
    _, thresh = cv2.threshold(blur, thresh=20, maxval=255, type=cv2.THRESH_BINARY)
    dilated = cv2.dilate(thresh, kernel=None, iterations=3)
    contours,_ = cv2.findContours(dilated, mode=cv2.RETR_TREE, method=cv2.CHAIN_APPROX_SIMPLE)
    
    # Graficar rectangulos
    for contour in contours:
        (x,y,w,h) = cv2.boundingRect(contour)
        if cv2.contourArea(contour) < 1000: # Si el area del contorno es muy chica, ignorar (omite contornos que no son personas)
            continue
        cv2.rectangle(frame1, pt1=(x,y), pt2=(x+w, y+h), color=(0,255,0), thickness=2)
        cv2.putText(frame1, text="Status: {}".format('Movement'), org=(10,20), fontFace=cv2.FONT_HERSHEY_SIMPLEX, 
                    fontScale=1, color=(0,0,255), thickness=3)
            
    #cv2.drawContours(frame1, contours, contourIdx=-1, color=(0,255,0), thickness=2)
    
    cv2.imshow('inter', frame1)
    frame1 = frame2
    ret, frame2 = cap.read()
                
    if cv2.waitKey(40) & 0xFF == ord('q'): # El 0xFF es una mascara (para sistema 64-bit) que se le pone al waitKey
        break # Si presiona 'q' sale del loop

cap.release()
cv2.destroyAllWindows()
```

### Detect Simple Geometric Shapes

```python
img = cv2.imread(path+'pic3.png') 
imggrey = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

_, thresh = cv2.threshold(imggrey, thresh=240, maxval=255, type=cv2.THRESH_BINARY)
contours,_ = cv2.findContours(thresh, mode=cv2.RETR_TREE, method=cv2.CHAIN_APPROX_SIMPLE)

for contour in contours:
    approx = cv2.approxPolyDP(contour, epsilon=0.01*cv2.arcLength(contour, closed=True), closed=True)
    cv2.drawContours(img, contours=[approx], contourIdx=0, color=(0,0,0), thickness=2)
    x = approx.ravel()[0]
    y = approx.ravel()[1]
    if len(approx) == 3: # Triangulo (3 puntos)
        x,
        cv2.putText(img, text='Triangle', org=(x,y), fontFace=cv2.FONT_HERSHEY_SIMPLEX, fontScale=0.5, color=(0,0,0))
    
    elif len(approx) == 4: # Rectangulo o Cuadrado
        x,y,w,h = cv2.boundingRect(approx)
        aspectRatio = float(w)/h
        if aspectRatio >= 0.95 and aspectRatio <= 1.05:
            cv2.putText(img, text='Square', org=(x,y), fontFace=cv2.FONT_HERSHEY_SIMPLEX, fontScale=1, color=(0,0,0))
        else:
            cv2.putText(img, text='Rectangle', org=(x,y), fontFace=cv2.FONT_HERSHEY_SIMPLEX, fontScale=1, color=(0,0,0))

    elif len(approx) == 5: # Pentagono
        cv2.putText(img, text='Pentagon', org=(x,y), fontFace=cv2.FONT_HERSHEY_SIMPLEX, fontScale=1, color=(0,0,0))
    
    elif len(approx) == 10: # Star
        cv2.putText(img, text='Star', org=(x,y), fontFace=cv2.FONT_HERSHEY_SIMPLEX, fontScale=1, color=(0,0,0))
        
    else:
        cv2.putText(img, text='Circle', org=(x,y), fontFace=cv2.FONT_HERSHEY_SIMPLEX, fontScale=1, color=(0,0,0))
        
cv2.imshow('shapes', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### Template Matching

```python
img = cv2.imread(path+'messi5.jpg')
grey_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
template = cv2.imread(path+'messi_face.jpg', 0)

res = cv2.matchTemplate(grey_img, templ=template, method=cv2.TM_CCOEFF_NORMED)
threshold = 0.8
loc = np.where(res >= threshold)

w, h = template.shape[::-1]

for pt in zip(*loc[::-1]):
    cv2.rectangle(img, pt1=pt, pt2=(pt[0]+w, pt[1]+h), color=(0,0,255), thickness=2)


cv2.imshow('image', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
